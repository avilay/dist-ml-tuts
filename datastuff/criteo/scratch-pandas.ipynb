```python
dense_feature_names = [f"I{i}" for i in range(1, 14)]
sparse_feature_names = [f"C{i}" for i in range(1, 27)]
colnames = ["label"] + dense_feature_names + sparse_feature_names
df = pd.read_csv(filepath, sep="\t", header=None, names=colnames)
```

```python
uniqs = {}
for col in sparse_feature_names:
    uniqs[col] = set(df[col].unique())
```
